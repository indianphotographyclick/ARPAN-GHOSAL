###  ***Descriptive Statistics & Visualization***

#### "How many participants have a family history of mental illness?"

```python
Family_History_Counts = data['Family  History'].value_counts()
Family_History_Counts
```

```python
plt.figure(figsize=(6, 6))
plt.pie(Family_History_Counts.values,labels=Family_History_Counts.index, autopct='%1.1f%%', colors=['skyblue', 'salmon'], startangle=90, explode=(0.05, 0))  
plt.title('Family History of Mental Illness (Pie Chart)')
plt.show()
```

#### What percentage of participants have sought mental health treatment?

```python
Treatment_Counts = data['Treatment'].value_counts()
Treatment_Percentages = Treatment_Counts / Treatment_Counts.sum() * 100
Treatment_Counts, Treatment_Percentages
```

```python
Treatment_Counts = data['Treatment'].value_counts()
plt.figure(figsize=(7, 5))
plt.bar(Treatment_Counts.index, Treatment_Counts.values, color=['red', 'navy'])
plt.title('Mental Health Treatment Seeking Behavior')
plt.xlabel('Sought Treatment')
plt.ylabel('Number of Participants')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### What is the gender distribution of participants?

```python
Gender_Counts = data['Gender'].value_counts()
Gender_Percentages = Gender_Counts / Gender_Counts.sum() * 100
Gender_Counts, Gender_Percentages
```

```python
plt.figure(figsize=(6, 6))
colors = ['skyblue', 'lightcoral', 'lightgreen', 'violet', 'gold']
explode = [0.05 if i == 0 else 0 for i in range(len(Gender_Counts))]
plt.pie(Gender_Counts.values, labels=Gender_Counts.index, autopct='%20.5f%%', colors=colors[:len(Gender_Counts)], startangle=90, explode=explode)
plt.title('Gender Distribution of Participants (Pie Chart)', fontsize=14)
plt.show()
```

### ***Advanced & Visualization***

#### How many people who did not receive treatment also reported struggling to cope?

```python
Struggling_Untreated = data[(data['Treatment'] == 'No') & (data['Coping Struggles'] == 'Yes')]
Count = Struggling_Untreated.shape[0]
Count
```

```python
Grouped = data.groupby(['Treatment', 'Coping Struggles']).size().unstack()
Grouped.plot(kind='bar', figsize=(8, 6), color=['lightgreen', 'lightcoral'])
plt.title('Coping Struggles vs Treatment Status (Grouped Bar Chart)', fontsize=14)
plt.xlabel('Treatment Received')
plt.ylabel('Number of Participants')
plt.legend(title='Coping Struggles')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### Among people who reported growing stress, how many also reported changes in their daily habits?

```python
Stress_And_Changes = data[(data['Growing Stress'] == 'Yes') & (data['Changes Habits'] == 'Yes')]
Count = Stress_And_Changes.shape[0]
Count
```

```python
Grouped = data.groupby(['Growing Stress', 'Changes Habits']).size().unstack()
Grouped.plot(kind='bar', figsize=(8, 6), color=['lightblue', 'salmon'])
plt.title('Changes in Habits by Growing Stress Status', fontsize=14)
plt.xlabel('Growing Stress')
plt.ylabel('Number of Participants')
plt.legend(title='Changes in Daily Habits')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### How many people who reported having social weakness also reported a drop in work interest?

```python
Social_Weakness = data[data['Social Weakness'] == 'Yes']
social_weak_and_disinterest = Social_Weakness[Social_Weakness['Work Interest'] == 'Yes'].shape[0]
f"People with social weakness and work interest drop: {social_weak_and_disinterest}"
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=data, x='Social Weakness', hue='Work Interest', palette='Set1')
plt.title("Relationship between Social Weakness and Work Interest Drop", fontsize=14)
plt.xlabel("Social Weakness")
plt.ylabel("Count")
plt.legend(title="Work Interest Drop")
plt.tight_layout()
plt.show()
```

#### Among those who have had a mental health interview, how many also received treatment?

```python
Interview_And_Treated = data[(data['Mental Health Interview'] == 'Yes') & (data['Treatment'] == 'Yes')]
Count = Interview_And_Treated.shape[0]
Count
```

```python
Interview_Group = data[data['Mental Health Interview'] == 'Yes']
Interview_Treatment_Counts = Interview_Group['Treatment'].value_counts()
plt.figure(figsize=(6, 6))
colors = ['lightgreen', 'lightcoral']
explode = [0.05 if i == 0 else 0 for i in range(len(Interview_Treatment_Counts))]
plt.pie(Interview_Treatment_Counts.values, labels=Interview_Treatment_Counts.index, autopct='%2.1f%%', colors=colors, startangle=90, explode=explode)
plt.title('Treatment Status Among Those Interviewed', fontsize=14)
plt.show()
```

#### Create a bar chart showing the number of respondents by gender. 

```python
Treatment = data.groupby(['Gender', 'Treatment']).size().reset_index(name='Count')
Treatment
```

```python
Treatment = data.groupby(['Gender', 'Treatment']).size().reset_index(name='Count')
plt.figure(figsize=(10, 6))
sns.barplot(data=Treatment, x='Gender', y='Count', hue='Treatment', palette='Set1')
plt.title('Mental Health Treatment by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Treatment')
plt.tight_layout()
plt.show()
```

#### Plot a bar chart of mental health treatment counts by country (Top 10 countries only).

```python
Country_Treatment_Counts = data.groupby(['Country', 'Treatment']).size().reset_index(name='Count')
Top_Countries = data['Country'].value_counts().head(10).index
Filtered_Data = Country_Treatment_Counts[Country_Treatment_Counts['Country'].isin(Top_Countries)]
plt.figure(figsize=(12, 6))
sns.barplot(data=Filtered_Data, x='Country', y='Count', hue='Treatment', palette='Set2')
plt.title('Mental Health Treatment by Country (Top 10)', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Number of Participants')
plt.xticks(rotation=45)
plt.legend(title='Treatment Status')
plt.tight_layout()
plt.show()
```

#### Create a grouped bar chart comparing treatment vs. no treatment across genders.

```python
Gender_Treatment = pd.crosstab(data['Gender'], data['Treatment'])
Gender_Treatment.plot(kind='bar', figsize=(10, 6), color=['#66b3ff', '#ff9999'])
plt.title('Mental Health Treatment by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Treatment')
plt.tight_layout()
plt.show()
```

#### Create a stacked bar chart to show the distribution of coping struggles by self-employment status.

```python
Coping_Vs_Self = pd.crosstab(data['Self Employed'], data['Coping Struggles'])
Coping_Vs_Self.plot(kind='bar', stacked=True, figsize=(8, 5), colormap='seismic')
plt.title('Coping Struggles by Self-Employment Status')
plt.xlabel('Self Employed')
plt.ylabel('Number of Respondents')
plt.legend(title='Coping Struggles')
plt.tight_layout()
plt.show()
```

#### Create a pie chart to show the percentage of people who have a family history of mental health issues.

```python
family_history_counts = data['Family  History'].value_counts()
plt.figure(figsize=(6, 6))
colors = ['#ffcc99', '#99ccff']
family_history_counts.plot.pie(autopct='%1.1f%%', startangle=140, colors=colors, labels=['Yes', 'No'])
plt.title('Family History of Mental Health Issues')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

#### Create a bar chart showing how many people spent different durations indoors.

```python
Grouped = data.groupby(['Days Indoors', 'Treatment']).size().reset_index(name='Count')
Grouped['Days Indoors'] = Grouped['Days Indoors'].astype(str)
plt.figure(figsize=(10, 6))
sns.barplot(data=Grouped, x='Days Indoors', y='Count', hue='Treatment', palette='Pastel2')
plt.title('Days Spent Indoors by Treatment Status')
plt.xlabel('Duration Indoors')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Treatment')
plt.tight_layout()
plt.show()
```

#### Create a horizontal bar chart showing the number of people who reported growing stress across employment types.

```python
Growing_Stress = data[data['Growing Stress'] == 'Yes']
Grouped = Growing_Stress.groupby(['Self Employed', 'Treatment']).size().reset_index(name='Count')
plt.figure(figsize=(8, 5))
sns.barplot(data=Grouped, x='Count', y='Self Employed', hue='Treatment', palette='Set1')
plt.title('People with Growing Stress by Employment Type and Treatment')
plt.xlabel('Number of Respondents')
plt.ylabel('Self Employed Status')
plt.legend(title='Treatment')
plt.tight_layout()
plt.show()
```

#### How does the number of people reporting “Growing Stress” vary across age groups or over time (if time data existed)?

```python
Trend = data.groupby(['Days Indoors', 'Growing Stress']).size().unstack().fillna(0)
Trend.plot(kind='line', figsize=(10, 5), marker='o')
plt.title('Trend of Growing Stress by Days Spent Indoors')
plt.xlabel('Days Indoors')
plt.ylabel('Number of Respondents')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Distribution of people by mood swings

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Mood Swings', hue='Gender', palette='Set2')
plt.title('Distribution of Mood Swings by Gender')
plt.xlabel('Mood Swings')
plt.ylabel('Number of Respondents')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

#### Is there a pattern between stress and habit change?

```python
Pattern = data.groupby(['Growing Stress', 'Changes Habits']).size().unstack()
Pattern.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='seismic')
plt.title('Pattern: Growing Stress vs Habit Changes')
plt.xlabel('Growing Stress')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
data["Glucose"]=data["Glucose"].replace(0,data["Glucose"].mean())
```

```python
plt.figure(figsize=(12,8),dpi=80)
plt.hist(data["Glucose"])
plt.xlabel("Glucose")
plt.ylabel("Count")
sns.set_style(style="darkgrid")
print("Mean of the Glucose is",data["Glucose"].mean())
print("Datatype of the Glucose is",data["Glucose"].dtypes)
```

```python
data["BloodPressure"]=data["BloodPressure"].replace(0,data["BloodPressure"].mean()) 
```

```python
plt.figure(figsize=(12,8),dpi=80)
plt.hist(data["BloodPressure"])
plt.xlabel("BloodPressure")
plt.ylabel("Count")
sns.set_style(style="darkgrid")
print("Mean of the BloodPressure is",data["BloodPressure"].mean())
print("Datatype of the BloodPressure is",data["BloodPressure"].dtypes)
```

```python
data["BMI"]=data["BMI"].replace(0,data["BMI"].mean()) 
```

```python
data["Insulin"]=data["Insulin"].replace(0,data["Insulin"].mean()) 
```

```python
plt.figure(figsize=(12,10),dpi=80)
plt.title("Checking missing values with heatmap")
sns.heatmap(data.isnull(),cmap="magma",yticklabels=False)
```

```python
plt.figure(figsize=(12,8),dpi=60)
plt.title("Count of Outcome")
plt.xlabel("Outcome")
plt.ylabel("Count")
sns.countplot(data["Outcome"])
sns.set_style(style="darkgrid")
print("Count of class:\n",data["Outcome"].value_counts())
```

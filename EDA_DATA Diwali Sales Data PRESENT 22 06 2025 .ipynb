### FILTERING 

##### 1. Which female customers from Maharashtra placed more than 2 orders?

```python
filtered_data = data[(data['Gender'] == 'F') & 
                     (data['State'] == 'Maharashtra') & 
                     (data['Orders'] > 2)]

filtered_data[['Cust_name', 'Orders', 'State', 'Gender']]
```

###### VISUALIZATION

##### 2. List customers aged below 25 who spent more than ₹20,000

```python
Customer = data[(data['Age'] < 25) & (data['Amount'] > 20000)]
Customer[['Cust_name', 'Age', 'Amount']]
```

##### 3. Find all IT Sector customers who ordered 'Auto' category products.

```python
Find = data[(data['Occupation'] == 'IT Sector') & 
            (data['Product_Category'] == 'Auto')]
Find[['Cust_name', 'Occupation', 'Product_Category']]
```

##### 4. Which states had total sales over ₹100,000?

```python
State = data.groupby('State')['Amount'].sum().reset_index()
State = State[State['Amount'] > 100000]
State
```

##### 5. Show all male customers from Southern zone with more than 1 order.

```python
Southern_Zone_Customers = data[(data['Gender'] == 'M') & 
                               (data['Zone'] == 'Southern') & 
                               (data['Orders'] > 1)]
Southern_Zone_Customers[['Cust_name', 'Gender', 'Zone', 'Orders']]
```

### VISUALIZATION

###### Product Category vs Sales

```python
CATEGORY_SALES = data.groupby('Product_Category')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=CATEGORY_SALES, x='Product_Category', y='Amount')
plt.title('Sales by Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###### Age Group vs Occupation by Total Amount

```python
PIVOT_TABLE = data.pivot_table(values='Amount', index='Occupation', columns='Age Group', aggfunc='sum', fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(PIVOT_TABLE, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Sales Heatmap: Age Group vs Occupation')
plt.tight_layout()
plt.show()
```

###### Top 10 States By Sales

```python
STATES = data.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
STATES.plot(kind='bar', color='teal')
plt.title('Top 10 States by Total Sales')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###### Gender-wise Participation

```python
plt.figure(figsize=(6, 6))
data['Gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Gender-wise Participation')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

###### Product Category vs Sales

```python
category_sales = data.groupby('Product_Category')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=category_sales, x='Product_Category', y='Amount')
plt.title('Sales by Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###### Zone-wise total Amount and total Orders

```python
Zone_Summary = data.groupby('Zone')[['Amount', 'Orders']].sum().reset_index().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=Zone_Summary.melt(id_vars='Zone'), x='Zone', y='value', hue='variable', palette='Set2')
plt.title('Zone-wise Total Orders & Total Amount')
plt.ylabel('Total Value')
plt.tight_layout()
plt.show()
```

###### Age vs Amount by Product Category

```python
Top_Categories = data['Product_Category'].value_counts().head(3).index
Filtered_Data = data[data['Product_Category'].isin(Top_Categories)]
plt.figure(figsize=(12, 6))
sns.histplot(
    data=Filtered_Data,
    x='Age',
    weights='Amount',
    hue='Product_Category',
    multiple='stack',
    bins=20,
    palette='tab10'
)
plt.title('Age Distribution by Purchase Amount in Top Product Categories')
plt.xlabel('Customer Age')
plt.ylabel('Total Purchase Amount')
plt.tight_layout()
plt.show()
```

##### 26) Compare intake condition counts by animal type (stacked bar plot).

```python
Condition = pd.crosstab(data['Animal Type'], data['Intake Condition'])
Condition.plot(kind='bar', stacked=True, figsize=(10,6), colormap='Set2')
plt.title('Intake Condition By Animal Type')
plt.xlabel('Animal Type'); plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

##### 28) Heatmap: Cross-tab of Animal Type vs Intake Type.

```python
Cross_Tab = pd.crosstab(data['Animal Type'], data['Intake Type'])
plt.figure(figsize=(8,5))
sns.heatmap(Cross_Tab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Animal Type vs Intake Type')
plt.tight_layout()
plt.show()
```

##### 32) Bar chart: Count of neutered vs intact animals grouped by Animal Type.

```python
status_type = pd.crosstab(data['Animal Type'], data['Neuter Status'])
status_type.plot(kind='bar', stacked=False, figsize=(10, 6), colormap='hsv')
plt.title('Neuter Status by Animal Type')
plt.xlabel('Animal Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

##### 33) Breed vs Intake Condition heatmap.

```python
Top10_Breeds = data['Breed'].value_counts().head(10).index
breed_cond = pd.crosstab(data[data['Breed'].isin(Top10_Breeds)]['Breed'], data['Intake Condition'])
plt.figure(figsize=(10,6))
sns.heatmap(breed_cond, annot=True, fmt='d', cmap='OrRd')
plt.title('Top 10 Breeds vs Intake Condition')
plt.tight_layout()
plt.show()
```

##### 34) Sex upon Intake vs Age histogram or KDE plot.

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=data[data['Sex upon Intake'].notna()], x='Age_in_Months', hue='Sex upon Intake', common_norm=False)
plt.title('Age Distribution by Sex upon Intake (KDE)')
plt.xlabel('Age (Months)')
plt.tight_layout()
plt.show()
```

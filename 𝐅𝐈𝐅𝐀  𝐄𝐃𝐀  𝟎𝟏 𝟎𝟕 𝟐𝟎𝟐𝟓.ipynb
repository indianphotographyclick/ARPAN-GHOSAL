### Visualization Questions 

#### 1. Bar chart of top 10 players by Value.

```python
Top_10_Value = data.nlargest(10, 'Value')[['Name', 'Value']]
plt.figure(figsize=(12, 6))
sns.barplot(data=Top_10_Value, x='Value', y='Name', palette='viridis')
plt.title('Top 10 Players by Value')
plt.xlabel('Value (in thousands €)')
plt.ylabel('Player Name')
plt.tight_layout()
plt.show()
```

#### 2. Histogram of player Ages.

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['Age'], bins=20, kde=False, color='teal')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.tight_layout()
plt.show()
```

#### 3. Boxplot of Wage distribution by Position.

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=data, x='Position', y='Wage', palette='Set3')
plt.xticks(rotation=90)
plt.title('Wage Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Wage (in thousands €)')
plt.tight_layout()
plt.show()
```

#### 4. Scatter plot of Value vs Overall rating.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Overall', y='Value', alpha=0.6, color='darkred')
plt.title('Value vs Overall Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Value (in thousands €)')
plt.tight_layout()
plt.show()
```

#### 5. Bar plot of average Overall by Club (top 10 clubs).

```python
Avg_Overall_By_Club = data.groupby('Club')['Overall'].mean().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=Avg_Overall_By_Club, x='Overall', y='Club', palette='coolwarm')
plt.title('Top 10 Clubs by Average Overall Rating')
plt.xlabel('Average Overall Rating')
plt.ylabel('Club')
plt.tight_layout()
plt.show()
```

#### 6. Pie chart of player distribution by Preferred Foot.

```python
Foot_Counts = data['Preferred Foot'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(Foot_Counts, labels=Foot_Counts.index, autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], startangle=140)
plt.title('Player Distribution by Preferred Foot')
plt.axis('equal')
plt.show()
```

#### 7. Line chart of average Potential by age.

```python
Avg_Potential_By_Age = data.groupby('Age')['Potential'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(data=Avg_Potential_By_Age, x='Age', y='Potential', marker='o', color='green')
plt.title('Average Potential by Age')
plt.xlabel('Age')
plt.ylabel('Average Potential')
plt.tight_layout()
plt.show()
```

#### 8. Boxplot of Height by Position.

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=data, x='Position', y='Height', palette='pastel')
plt.xticks(rotation=90)
plt.title('Height Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Height (in feet)')
plt.tight_layout()
plt.show()
```

#### 9. Correlation heatmap of numerical features.

```python
Numerical_Cols = data.select_dtypes(include=['float64', 'int64'])
Correlation_Matrix = Numerical_Cols.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(Correlation_Matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```

#### 10. Distribution plot of Release Clause.

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['Release Clause'], bins=50, kde=True, color='coral')
plt.title('Distribution of Release Clause')
plt.xlabel('Release Clause (in thousands €)')
plt.ylabel('Number of Players')
plt.tight_layout()
plt.show()
```

#### 11. Count plot of players by Nationality (top 10 only).

```python
Top_10_Nations = data['Nationality'].value_counts().head(10).index
Top_10_Nations_2 = data[data['Nationality'].isin(Top_10_Nations)]
plt.figure(figsize=(12, 6))
sns.countplot(data=Top_10_Nations_2, y='Nationality', order=Top_10_Nations, palette='magma')
plt.title('Top 10 Nationalities by Player Count')
plt.xlabel('Number of Players')
plt.ylabel('Nationality')
plt.tight_layout()
plt.show()
```

#### 12. Pair plot of Wage, Value, Overall, Potential.

```python
pairplot_data = data[['Wage', 'Value', 'Overall', 'Potential']]
sns.pairplot(pairplot_data, diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Wage, Value, Overall, and Potential', y=1.02)
plt.show()
pairplot_data_clean = data[['Wage', 'Value', 'Overall', 'Potential']].apply(pd.to_numeric, errors='coerce').dropna()
sns.pairplot(pairplot_data_clean, diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Wage, Value, Overall, and Potential', y=1.02)
plt.show()
```

#### 13. Horizontal bar chart of average Weight by Position.

```python
Avg_Weight_By_Position = data.groupby('Position')['Weight'].mean().sort_values()
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_weight_by_position.values, y=avg_weight_by_position.index, palette='cubehelix')
plt.title('Average Weight by Position')
plt.xlabel('Average Weight (lbs)')
plt.ylabel('Position')
plt.tight_layout()
plt.show()
```

#### 14. Heatmap of club vs number of players (top 10 clubs).

```python
top_10_clubs = data['Club'].value_counts().head(10)
club_counts_df = pd.DataFrame(top_10_clubs).reset_index()
club_counts_df.columns = ['Club', 'Player Count']
club_counts_df = club_counts_df.set_index('Club')
plt.figure(figsize=(8, 6))
sns.heatmap(club_counts_df, annot=True, fmt='d', cmap='YlGnBu', cbar=False)
plt.title('Number of Players per Top 10 Club')
plt.xlabel('Player Count')
plt.ylabel('Club')
plt.tight_layout()
plt.show()
```

#### 15. KDE plot of Height for GKs vs non-GKs.

```python
gks = data[data['Position'] == 'GK']
non_gks = data[data['Position'] != 'GK']
plt.figure(figsize=(10, 6))
sns.kdeplot(gks['Height'], label='Goalkeepers', shade=True, color='blue')
sns.kdeplot(non_gks['Height'], label='Non-Goalkeepers', shade=True, color='orange')
plt.title('KDE Plot of Height: GKs vs Non-GKs')
plt.xlabel('Height (in feet)')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```

#### 16. Plot average Value per year joined (trend over years).

```python
data['Joined'] = pd.to_datetime(data['Joined'], errors='coerce')
data['JoinYear'] = data['Joined'].dt.year
avg_value_by_year = data.groupby('JoinYear')['Value'].mean().dropna()
plt.figure(figsize=(10, 6))
sns.lineplot(x=avg_value_by_year.index, y=avg_value_by_year.values, marker='o', color='purple')
plt.title('Average Player Value by Year Joined')
plt.xlabel('Join Year')
plt.ylabel('Average Value (in €)')
plt.tight_layout()
plt.show()
```

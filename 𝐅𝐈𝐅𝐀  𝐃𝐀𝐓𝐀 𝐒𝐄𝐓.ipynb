### ***4. Visualization*** 

#### **Bar Chart Of Top 10 Players By Value.**

```python
Top_10_Value = data.nlargest(10, 'Value')[['Name', 'Value', 'Position']]
plt.figure(figsize=(12, 6))
sns.barplot(data=Top_10_Value, x='Value', y='Name', hue='Position', dodge=False, palette='viridis')
plt.title('Top 10 Players By Value (Grouped By Position)', fontsize=14, fontweight='bold')
plt.xlabel('Value (in thousands €)')
plt.ylabel('Player Name')
plt.tight_layout()
plt.show()
```

#### **Histogram Of Player Ages.**

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['Age'], bins=20, kde=False, color='teal')
plt.title('Distribution Of Player Ages')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.tight_layout()
plt.show()
```

#### **Scatter Plot Of Value Vs Overall Rating.**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Overall', y='Value', alpha=0.6, color='darkred')
plt.title('Value Vs Overall Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Value (in thousands €)')
plt.tight_layout()
plt.show()
```

#### **Pie Chart Of Player Distribution By Preferred Foot.**

```python
Foot_Counts = data['Preferred Foot'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(Foot_Counts, labels=Foot_Counts.index, autopct='%10.5f%%', colors=['#66b3ff','#ff9999'], startangle=140)
plt.title('Player Distribution By Preferred Foot')
plt.axis('equal')
plt.show()
```

#### **Line Chart Of Average Potential By Age.**

```python
Avg_Potential_By_Age = data.groupby('Age')['Potential'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(data=Avg_Potential_By_Age, x='Age', y='Potential', marker='o', color='green')
plt.title('Average Potential by Age')
plt.xlabel('Age')
plt.ylabel('Average Potential')
plt.tight_layout()
plt.show()
```

#### **Boxplot Of Height By Position.**

```python
data['Position Type'] = data['Position'].apply(lambda x: 'Goalkeeper' if x == 'GK' else 'Non-Goalkeeper')
plt.figure(figsize=(14, 6))
sns.boxplot(data=data, x='Position', y='Height', hue='Position Type', palette="pastel")
plt.xticks(rotation=90)
plt.title('Height Distribution By Position (Grouped By Position Type)', fontsize=14, fontweight='bold')
plt.xlabel('Position')
plt.ylabel('Height (in feet)')
plt.tight_layout()
plt.show()
```

#### **Correlation Heatmap Of Numerical Features.**

```python
Numerical_Cols = data.select_dtypes(include=['float64', 'int64'])
Correlation_Matrix = Numerical_Cols.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(Correlation_Matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.title('Correlation Heatmap Of Numerical Features')
plt.tight_layout()
plt.show()
```

#### **Count Plot Of Players By Nationality (Top 10 only).**

```python
data['Position Type'] = data['Position'].apply(lambda x: 'Goalkeeper' if x == 'GK' else 'Non-Goalkeeper')
Top_10_Nations = data['Nationality'].value_counts().head(10).index
Top_10_Nations_2 = data[data['Nationality'].isin(Top_10_Nations)]
plt.figure(figsize=(12, 6))
sns.countplot(data=Top_10_Nations_2, y='Nationality', order=Top_10_Nations, hue='Position Type', palette='magma')
plt.title('Top 10 Nationalities By Player Count (Grouped By Position Type)', fontsize=14, fontweight='bold')
plt.xlabel('Number of Players')
plt.ylabel('Nationality')
plt.tight_layout()
plt.show()
```

#### **Pair Plot Of Wage, Value, Overall, Potential.**

```python
pairplot_data = data[['Wage', 'Value', 'Overall', 'Potential']]
sns.pairplot(pairplot_data, diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Wage, Value, Overall, and Potential', y=1.02)
plt.show()
pairplot_data_clean = data[['Wage', 'Value', 'Overall', 'Potential']].apply(pd.to_numeric, errors='coerce').dropna()
sns.pairplot(pairplot_data_clean, diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Wage, Value, Overall, and Potential', y=1.02)
plt.show()
```

####  **KDE Plot Of Height For GKs Vs Non-GKs.**

```python
data['Position Type'] = data['Position'].apply(lambda x: 'Goalkeeper' if x == 'GK' else 'Non-Goalkeeper')
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x='Height', hue='Position Type', fill=True, palette=['blue', 'orange'])
plt.title('KDE Plot of Height: Goalkeepers Vs Non-Goalkeepers')
plt.xlabel('Height (in feet)')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```
